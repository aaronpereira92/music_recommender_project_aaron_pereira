```python
#1. import
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#2. url stored in variable
url = "https://www.xxlmag.com/hip-hop-songs-defined-2010s/"
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
# 4. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
#5. Went onto the xxlmag website, and chose the appropriate selector to get the music info
soup.select("#post-874198 > div.content-wrap > div.content > div > div.list-post.standard.media-bottomx > ul > li:nth-child(2) > div > div > header > h2")
```

```python
#soup.select("div > ul > li > div > div > header > h2")
modern_rap_songs = []
for title in soup.select("div > ul > li > div > div > header > h2"):
    modern_rap_songs.append(title.get_text().replace('"','').replace('“','').replace('”',''))
```

```python
modern_rap_artists = []
for title in soup.select("div > ul > li > div > div > header > small"):
    modern_rap_artists.append(title.get_text())
```

```python
#modern rap dataframe
modern_rap = pd.DataFrame({"song": modern_rap_songs, "artist": modern_rap_artists})
```

```python
modern_rap
```

```python
#500 greatest songs of all time from Rolling stones magazine
url = "https://stuarte.co/2021/2021-full-list-rolling-stones-top-500-songs-of-all-time-updated/"
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
soup.select("#post-14376 > div.entry-content > div.table-responsive > table > tbody > tr > td")
```

```python
rs_artists = []
for index, elem in enumerate(soup.select("#post-14376 > div.entry-content > div.table-responsive > table > tbody > tr > td")):
    if ( index%2 == 0) and (index != 0):
        rs_artists.append(elem.get_text())
rs_artists_new = []
for index, elem in enumerate(rs_artists):
    if ( index%2 == 0):
        rs_artists_new.append(elem.strip())
```

```python
rs_artists_new
```

```python
rs_songs= []
for index, elem in enumerate(soup.select("#post-14376 > div.entry-content > div.table-responsive > table > tbody > tr > td")):
    if ( index%2 != 0) and (index != 0):
        rs_songs.append(elem.get_text())
rs_songs_new = []
for index, elem in enumerate(rs_songs):
    if ( index%2 == 0):
        rs_songs_new.append(elem.strip())
```

```python
rs_songs_new
```

```python
rs_500 = pd.DataFrame({"song": rs_songs_new, "artist": rs_artists_new})
```

```python
rs_500
```

## Importing music dataset from kaggle, which contains about 1.2 million songs
Will take a **random subset** of about 10,000 songs which will be use in addition to the RS 500 and moderrap songs 

```python
music_data = pd.read_csv('tracks_features.csv')
```

```python
pd.get_option("display.max_columns")
music_data
```

```python
music_data_short = music_data.sample(frac=0.004, replace=False, random_state=10)
```

```python
music_data_short
```

```python
music_data_short.columns
```

```python
music_data_final = music_data_short[['name', 'artists']].reset_index(drop = True)
```

```python
music_data_final
```

```python
artists = []
for x in music_data_final['artists']:
    artists.append(x.replace('[','').replace(']',''))
artists
```

```python
artistdf = pd.DataFrame({'artist': artists})
artistdf
```

```python
artistdf = artistdf.replace({'\'': ''}, regex=True)
```

```python
artistdf
```

```python
music_data_final_2 = pd.concat([music_data_final,artistdf], axis =1)
```

```python
music_data_final_2.drop('artists',axis=1, inplace=True)
```

```python
music_data_final_2 = music_data_final_2.rename(columns={'name': 'song'})
music_data_final_2
```
